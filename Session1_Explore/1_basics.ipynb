## Object/Variable Assignment  
Values are assigned to an object using the "=" operator. It is worth noting that assigning a value to an object creates a reference to the object. It does not actually copy the object. A few more noteworthy aspects of variables in Python are  

- The variable or object is always on the left of the assignment operator.  
- Python evaluates the information to the right of the assignment operator first.  
- If the Python shell is reset, all object values are lost.  

```python
a = 45
b = 34
a * b
# 45 = c # will generate an error
```

Python evaluates information to the right of the assignment operator first.

### Typecasting  
- Process of converting one type to another.  
- Done explicitly by using type functions such as str(), int(). In the language R, this is referred to as coercion.   
- Occurs implicitly when Python encounters an operation that can only be carried out with a different type  

```python
str(7)
int(7.777)
float(7)
bool(7)
bytes(7)
```

Arithmetic operations cannot be carried out with bool type, so these are converted to int before performing addition. 

```python
object = True + True
isinstance(object, int)
```

When creating a variable, one does not have to specify the type. In this respect Python is like R, not Java. Python uses strong type which means once a variable is assigned a value, its type is enforced and cannot be automatically or implicitly converted to another type without an explicit cast or operation.

## Binary Operators
Python has a set of binary operators like in most other languages and use familiar math notation. These work as expected with numbers, however they can also work with string types yielding handy results. 

```python
12 + 5
12 - 5
12 * 5
12 / 5
"I am " + " smart"
"I am smart." * 5
```

### Common binary operations. 

```python
a = 1
b = 2
a + b # Addition
a - b # Subtraction
a * b # Multiplication
a/b   # Division
a**b  # Power function
a//b   # Floor division
a % b  # Module operator for remainder
(a>1) & (b>1) # True if both conditions are met
(a>1) | (b>1) # True if either condition is met
(a>1) ^ (b>1) # True only if either condition is met but not both; like XOR
a == b # Returns True if condition is met
a != b  # Returns True if condition is met
a < b # Returns True if condition is met
a > b # Returns True if condition is met
a is b # True if a and b reference same object
a is not b # True if a and b reference different objects
```

### Creating a Function

Python and its libraries have built-in functions that perform almost all operations desired by a data scientist. There are two reasons to create a function:    

- There is no function to perform the desired task  
- A set of tasks are performed repeatedly (say more than three times)  

To illustrate, we will create a function that will convert temperature from C to F  

```python
def convert_temp_c_to_f(temp_c = 37): # default value is set
    temp_f = temp_c * 9/5 + 32
    return(str(temp_c) + "C is " + str(temp_f) + "F")
```

Now, we invoke the function for temperature of 25C.

```python
convert_temp_c_to_f(temp_c= 25)
```

Here is another example

```python
def square_num(x):
    return(x**2)
square_num(45)
```

Python also supports creating functions on the fly, known as anonymous functions or lambda. In certain instances, this can reduce the amount of code required. 

```python
square_lambda =  lambda x:x**2
square_lambda(45)
```

## Method  

- A method is like a function that belongs to an object  
- Like a function, a method can take one or more arguments  
- To illustrate, we will examine methods for a *string* object.

```python
sentence = "  Nikhil is a good kid   "
```

```python
sentence.upper()
sentence.title()
sentence.capitalize()
sentence.swapcase()
sentence.strip()
sentence.lstrip()
sentence.rstrip()
sentence.strip().title()  # chaining methods
sentence.strip().replace("kid","boy")
sentence.startswith("R")
```

## Attributes  
- While methods are functions associated with an object, attributes are characteristics of the object   
- Attributes are Python objects stored inside the object   
- While attributes are called using the dot notation like methods, they do not have parentheses. E.g., object.my_method() vs. object.my_attribute
