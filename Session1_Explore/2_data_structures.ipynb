## Tuple
A tuple is a fixed-length immutable sequence of Python objects. Tuple can be created using parentheses or invoking tuple(). That said, the parentheses for creating a tuple is a norm, not a requirement. 

```python
my_tuple1 = (1,2,3)
```

```python
my_tuple2 = tuple([4,5,6])
type(my_tuple2)
```

Elements of a tuple can be accessed using square brackets

```python
my_tuple1[0:2] 
```

Tuples can be concatenated using the + operator

```python
(1,2,3) + (4,5,6)
```

Tuple Methods: Tuples do not have too many methods. A particularly useful method is count which counts occurrences of a value. 

```python
('Data', 'Data', 'Analysis', 'AI', 'Data').count('Data')
```

## List  

- A *list* is a data structure that holds an ordered collection of values.  
- Unlike tuples, lists are mutable and have variable length.  
- Although a list can contain elements of different types, lists generally have elements of the same type.   
- We often use the term *element* to describe an item in a list.  
- The length of a list is its number of elements.   
- A list may be instantiated using [ ] or list( ).  


```python
lst1 = [1,2,3,4]
lst1
```

```python
lst2 = list((5,6,7,8))
lst2
```

Like a tuple, a list can be subset using a position index and two lists can be concatenated using the + operator. 

```python
lst1[1:3] + lst2[1:3]
```

list() is frequently used in data processing to materialize an iterator or generator expression

```python
range(10,100,10)  # this will not display the range elements
list(range(10, 100, 10))
```

#### Functions
Functions like len() and type() can be used to learn more about the list

```python
len(lst1)
type(lst1)
```

#### Methods  

Lists have a number of methods that can be used to mutate or change the list object. Since lists are mutable, after the method is invoked, the result does not have to be assigned to an object. Here are a set of common methods for a list called lst:   

- lst.append(): Add an element to the end  
- lst.remove(): Remove a specific element  
- in: Check if an element is present in the list  
- not in: Check if an element is not in the list  
- lst.extend(): Add elements to the end of the list  
- lst.pop(): Removes last element in the list  
- lst.sort(): Arrange elements
  

```python
lst = ['a','b','c','d']
lst.append('e')             # Add an elements to the end
lst.remove('a')             # Remove a specific element
'a' in lst                  # Check if an element is present in the list
'a' not in lst              # Check if an element is not in the list
lst.extend(['f','g'])       # Add elements to the end of the list
lst.sort(reverse = True)    # Arrange elements
```

#### Slicing  

- Since a list holds an ordered collection of values, elements can be identified by their position.   
- Each element is assigned an index position.   
- Indices are consecutive and begin in 0.   
- Square brackets are used for extracting an element by index position.   
- A negative sign before an index means values are extracted relative to the end of the list.  

```python
lst
lst[0]
lst[1]
lst[-2]
```

- A start:stop notation can be passed to a list to subset or slice multiple elements. In this notation, the first element is inclusive but the last element is not included.

```python
lst[1:3]
```

- A missing index on either side of the : in a slicing operation is a shortcut to indicate all elements before or after. 

```python
lst[3:6]   # same result as the next slicing operation. 
lst[3:]    # all elements after 3
lst[-2:]   # all elements from second last to end
lst[::3]   # Every third element
```

It is possible to assign values to specified index positions in a list.

```python
print(lst)
lst[1:3]  = ['x','x']
print(lst)
```

## Dictionary  

- A dictionary is a data structure that stores data as a collection of key-value pairs, where the key and value are Python objects. The key can be used to identify a value to be retrieved, inserted, modified or deleted. A key is a unique identifier for a value. A value corresponds to the key. The values can contain duplicates, the keys should not.    
- One way to distinguish a dictionary from a list is to recognize that a dictionary is used to create an association, while a list is defined to place elements in order.   
- A Python dictionary is similar to the Webster Dictionary where words are associated to their meaning. One can lookup a word in the dictionary and find its corresponding meaning. There are many real-world situations where data includes a set of key-value pairs such as menu item and price on a restaurant menu, customer and dollars spent in a customer transaction file, etc.   
- One way to construct a dictionary is using curly braces.   

```python
empty_dict = {}
dict1 = {"name": "Rohan", 
         "age": 14, 
         "address": "123 Main Street, Gotham City, NY"} 
dict1                     # show contents of dictionary
```

```python
dict1['name']             # Extract value of the key name
dict1['gender'] = 'Male'  # Insert a new key-value pair by defining them
"name" in dict1           # Check for the presence of a key in the dictionary
"Rohan" in dict1.values() # Search through all values
dict1['customer'] = 'yes' # Insert another key-value pair, Dictionaries are mutable
del dict1['customer']     # delete or remove
dict1['customer'] = 'yes' # insert again to pop() in the next line
dict1.pop('customer')     # delete or remove and show deleted value
list(dict1.keys())        # Extract all keys
list(dict1.values())      # Extract all values
list(dict1.items())       # Extract all key-value pairs
dict2 = dict1           
dict2.update({'name': 'Nikhil',             # update dictionary
             'age':12,
             'address': '123 Main Street, Gotham City, NY',
             'gender': 'Male'})
```

## Set
A set is an unordered collection of unique elements. A set can be created using the set() or curly braces

```python
set([1,2,3,4,5])     # Convert a list to a set using set()
{1,2,3,4,5}          # Or use braces
a = {1,2,3,4,5,6}
b = {3,4,5,6,7,8}
a.union(b)         #  Sets support mathematical set operations like union, intersection, difference, and symmetric difference.
a.intersection(b)  # or a&b
a&b                # intersection
```

## Class    
* A class is a blue print for creating objects.  
* A class defines the methods/functionalities that objects made from it will have  
* An object is called an instance of the class it is made from  
* The act of creating an object is called instantiation  
* Every time we work with an Python object, it has been "instantiated" from a Class.   
* Syntax options like ""  or [] or {} are shortcuts for instantiation  
* For other classes, we will need to use class() syntax to instantiate an object
